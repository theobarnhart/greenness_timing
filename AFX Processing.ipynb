```python
# calculate ET from LE
# uses T in degrees C
# uses LE in W/m2
def le2et(x):
    
    cond = 2500.8-(2.36*x.TA)+(0.0016*(x.TA**2))+(0.00006*(x.TA**3)) #J/g/m2 latent heat as a fxn of temperature
    
    ET = x.LE/cond # calculate ET in g/m2
    
    #this might not be necessary
    if x.TA <= 0:
        ET = ET/916.7 # kg/m3 Density of ice [g/m2 -> m3/m2]
    else:
        ET = ET/1000.0 # kg/m3 # density of water [g/m2 -> m3/m2]

        
    ET = ET/1000 # m/m2
    
    ET = ET*1000 # mm/m2    
    
    return ET
```

```python
def wflux2et(H2O):
    mmass = 18.01528 # [g/mol]
    rho_water = 999.97 # [kg/m3]
    rho_water *= 1000 # kg -> g
    
    # convert the mmols to mols
    H2O /= 1000.
    
    mass = mmass*H2O # get the mass of the water
    
    # get the volume of water
    vol = mass/rho_water # m3
    depth = vol*1000
    
    return depth
```

# Calculate some Metrics from the Data

```python
years = unique(afx.index.year) # get all the years we have data for
```

```python
wyyears = years[1:] # get the water years we have data for
```

```python
afx3 = afx.resample('D',how=np.sum) # resample to daily sums 
```

```python
n = len(afx3)
afx3['cumulative_ET'] = np.zeros([n,1])
afx3['WY'] = np.zeros([n,1])
afx3['DOWY'] = np.zeros([n,1])

wystart = '-10-02' 
wyend = '-10-01'

for year in wyyears:
    
    afx3.loc[str(year-1)+wyend,'cumulative_ET'] = afx3.loc[str(year-1)+wyend].tot_ET
    afx3.loc[str(year-1)+wyend,'cumulative_NEE'] = afx3.loc[str(year-1)+wyend].tot_NEE
    afx3.loc[str(year-1)+wyend,'cumulative_GPP'] = afx3.loc[str(year-1)+wyend].tot_GPP
    
    afx3.loc[str(year-1)+wyend,'WY'] = year
    afx3.loc[str(year-1)+wyend,'DOWY'] = 1
    
    DOWY = 1
    
    bar = afx3.loc[str(year-1)+wyend].cumulative_ET # pull out the value for the subsequent interations
    bar2 = afx3.loc[str(year-1)+wyend].cumulative_NEE
    bar3 = afx3.loc[str(year-1)+wyend].cumulative_GPP
    
    for day in pd.date_range(str(year-1)+wystart,str(year)+wyend,freq='D'):
        day = str(day)
        afx3.loc[day,'cumulative_ET'] = afx3.loc[day].tot_ET + bar
        afx3.loc[day,'cumulative_NEE'] = afx3.loc[day].tot_NEE + bar2
        afx3.loc[day,'cumulative_GPP'] = afx3.loc[day].tot_GPP + bar3        
        
        bar = afx3.loc[day].cumulative_ET
        bar2 = afx3.loc[day].cumulative_NEE
        bar3 = afx3.loc[day].cumulative_GPP
        
        afx3.loc[day,'WY'] = year 
        
        DOWY += 1 # add one to the day of water year
        
        afx3.loc[day,'DOWY'] = DOWY
```

```python
afx3.cumulative_NEE /= 1000
afx3.cumulative_NEE /= 100
afx3.cumulative_GPP /= 1000
afx3.cumulative_GPP /= 100
```
