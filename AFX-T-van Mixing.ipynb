# Mixing AFX and T-Van ET into one signal

```python
%matplotlib inline
import numpy as np
import pandas as pd
import pylab
import matplotlib.pyplot as plt
import scipy as sp
import glob
from mpl_toolkits.mplot3d import Axes3D

pylab.rcParams['figure.figsize'] = (22.0, 11.0)
```

## Bring in T-Van ET

```python
tvan = pd.read_pickle('./ameriflux/t_van/tvan_et.pcl')
```

## Bring in AFX ET

```python
afx = pd.read_pickle('./ameriflux/niwot/AFX_niwot.pcl')
```

### Combine the T-van and AFX ET observations based on the proportion of the basin that they cover.

### 31% of the catchment is above treeline

```python
alp_frac = 0.31
```

```python
afx.describe()
```

```python
tvan.describe()
```

```python
ET=pd.DataFrame(np.zeros([len(tvan),1]),columns=['et'],index=tvan.index) 
```

```python
 ET.et = afx.tot_ET[tvan.index]*(1-alp_frac) + tvan.ET*(alp_frac)
```

```python
plt.plot(afx['2007-01-01':'2012-12-31'].index, afx['2007-01-01':'2012-12-31'].tot_ET, color = 'b', alpha = 0.3, lw=2);
plt.plot(ET.index, ET.et, color = 'c', alpha = 0.3, lw=2)
plt.plot(tvan[:'2012-12-31'].index, tvan[:'2012-12-31'].ET, color = 'r', alpha = 0.3, lw=2);
plt.legend(['AFX','T-van','Mix'], fontsize=18);
plt.axhline(0,c='w')
plt.title('Como Creek Vapor Fluxes', fontsize=24)
plt.ylabel('Vapor Flux [mm]', fontsize = 18);
```
