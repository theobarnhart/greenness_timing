```python
%matplotlib inline

import numpy as np
import pandas as pd
import pylab
import matplotlib.pyplot as plt
import scipy as sp
import glob

pylab.rcParams['figure.figsize'] = (22.0, 11.0)
```

```python
niwot = pd.read_pickle('./data/AMS_SWE_from_depth_WY_2006_2014.pcl') # SWE in cm
```

```python
P = pd.read_pickle('./data/boulder_co.pcl')
```

```python
niwot.swe *= 10. # conver to mm
```

```python
years = np.unique(niwot.index.year)
print 'Calendar Years: '+ str(years)
wyears = years[1:]
print 'Water Years: ' + str(wyears)
```

```python
cols = ['peakswe','peakswedate','snowfreedate','ablationrate','ablationseason', 'growseasonlength','ablationseasonP',
        'ablationseasonRain','ablationseasonSnow','ablseas_total_P']
SWE = pd.DataFrame(np.zeros([len(wyears),len(cols)]), index=wyears, columns=cols)
```

```python
SWE
```

```python
for year in wyears:
    
    SWE.loc[year,'peakswe'] = niwot.loc[
str(year-1)+'-12-01':str(year)+'-06-24'].swe.max() # extract the peak swe for the water year selected
    
    SWE.loc[year,'peakswedate'] = niwot.loc[
str(year-1)+'-12-01':str(year)+'-06-24'].swe.argmax() # extract the date of peak swe
    
    SWE.loc[year,'snowfreedate'] = niwot.loc[
str(year)+'-01-01':str(year)+'-06-24'].swe.argmin() # extract the first snow free date
    
    timedelta = SWE.loc[year,'snowfreedate'] - SWE.loc[year,'peakswedate']

    SWE.loc[year,'ablationrate'] = SWE.loc[year,'peakswe']/timedelta.days # [mm/day]
    SWE.loc[year, 'ablationseason'] = int(timedelta.days)
    
    timedelta = pd.datetime.strptime(str(year)+'-09-30','%Y-%m-%d') - SWE.loc[year,'peakswedate']
    
    strt = SWE.loc[year,'peakswedate']
    nd = SWE.loc[year,'snowfreedate']
    
    SWE.loc[year,'ablationseasonP'] = P[strt:nd].P.sum()/SWE.loc[year,'ablationseason'] # compute and store avearge daily precipitation during the melt season
    
    SWE.loc[year,'ablationseasonRain'] = P[strt:nd].rain.sum()
    SWE.loc[year,'ablationseasonSnow'] = P[strt:nd].snow.sum()
    
    SWE.loc[year,'ablseas_total_P'] = P[strt:nd].P.sum()
    
    SWE.loc[year,'growseasonlength'] = timedelta.days
    
SWE['flux'] = SWE.ablationrate+SWE.ablationseasonP
```
