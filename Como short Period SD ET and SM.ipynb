```python
yrs = np.unique(P.index.year)
wyrs = yrs[3:-2]
```

```python
wyrs
```

```python
cols = ['peakswe','peakswedate','snowfreedate','ablationrate','ablationseason', 'growseasonlength', 'ablseasonP']
SWE = pd.DataFrame(np.zeros([len(wyrs),len(cols)]), index=wyrs, columns=cols)
```

```python
SWE
```

```python
for year in wyrs:
    SWE.loc[year,'peakswe'] = niwot.loc[
str(year-1)+'-11-01':str(year)+'-09-30'].swe.max() # extract the peak swe for the water year selected
    
    SWE.loc[year,'peakswedate'] = niwot.loc[
str(year-1)+'-11-01':str(year)+'-09-30'].swe.argmax() # extract the date of peak swe
    
    SWE.loc[year,'snowfreedate'] = niwot.loc[
str(year)+'-01-01':str(year)+'-09-30'].swe.argmin() # extract the first snow free date
    
    timedelta = SWE.loc[year,'snowfreedate'] - SWE.loc[year,'peakswedate']

    SWE.loc[year,'ablationrate'] = SWE.loc[year,'peakswe']/timedelta.days # [mm/day]
    SWE.loc[year, 'ablationseason'] = int(timedelta.days)
    
    timedelta = pd.datetime.strptime(str(year)+'-09-30','%Y-%m-%d') - SWE.loc[year,'peakswedate']
    
    strt = SWE.loc[year,'peakswedate']
    nd = SWE.loc[year,'snowfreedate']
    
    SWE.loc[year,'ablseasonP'] = P[strt:nd].P.sum()/SWE.loc[year,'ablationseason'] # compute and store avearge daily precipitation during the melt season
    
    SWE.loc[year,'growseasonlength'] = timedelta.days
```

```python
SWE['water_flux'] = SWE.ablationrate + SWE.ablseasonP
```

```python
SWE
```

```python
# preallocate an output data frame
cols = ['melt','rain','flux','ET','del_sm0','del_sm1','del_sm2','del_sm3','sm'] # sm0-3 cooresponds to a lag of 0-3 days
data = pd.DataFrame(columns=cols)

td = pd.datetools.timedelta(1)
tdd = pd.datetools.timedelta(2)
tddd = pd.datetools.timedelta(3)
tdddd = pd.datetools.timedelta(4)

for year in wyrs: # loop through water years
    
    # grab the start and end of the melt season for each water year
    strt = SWE.loc[year,'peakswedate']
    end = SWE.loc[year,'snowfreedate']+pd.datetools.timedelta(10)
    
    dat = niwot[strt:end].copy() # subset the SWE data
    pdat = P[strt:end].copy() # subset the CRN P data
    
    dat['melt'] = np.zeros([len(dat),1]) # preallocate
    dat['melt'] = np.NaN
    dat['rain'] = np.NaN
    dat['flux'] = np.NaN
    
    for day in dat.index[1:]:
        dat.loc[day,'melt'] = dat.loc[day-td,'swe']-dat.loc[day,'swe'] # calculate the melt each day
        dat.loc[day,'rain'] = pdat.loc[day,'rain']
        dat.loc[day,'swe'] = dat.loc[day,'swe']
    
    dat['flux'] = dat.melt+dat.rain # sum precip and melt to get total water flux [mm/d]
    
    tmp=pd.DataFrame(dat[['melt','rain','flux','swe']]) # create a temporary data frame to hold the calculated results
    del dat
    del pdat
    
    dat = afx[strt:end] # subset the afx data
    SMdat = dat.resample('D',how=np.nanmean) # resample soil moisture to daily data
    ETdat = dat.resample('D',how=np.nansum) # resample ET to daily data

    # preallocate 
    SMdat['del_sm0'] = np.zeros([len(SMdat),1])
    SMdat['del_sm0'] = np.NaN
    SMdat['del_sm1'] = SMdat['del_sm0']
    SMdat['del_sm2'] = SMdat['del_sm0']
    SMdat['del_sm3'] = SMdat['del_sm0']
    
    for day in SMdat.index[1:-4]:
        SMdat.loc[day,'del_sm0'] = SMdat.loc[day,'SWC1']-SMdat.loc[day-td,'SWC1']
        SMdat.loc[day,'del_sm1'] = SMdat.loc[day+td,'SWC1']-SMdat.loc[day,'SWC1']
        SMdat.loc[day,'del_sm2'] = SMdat.loc[day+tdd,'SWC1']-SMdat.loc[day+td,'SWC1']
        SMdat.loc[day,'del_sm3'] = SMdat.loc[day+tddd,'SWC1']-SMdat.loc[day+tdd,'SWC1']
        
        
    # fill the output dataset
    tmp['ET'] = ETdat['tot_ET']
    tmp['del_sm0'] = SMdat['del_sm0']
    tmp['del_sm1'] = SMdat['del_sm1']
    tmp['del_sm2'] = SMdat['del_sm2']
    tmp['del_sm3'] = SMdat['del_sm3']
    tmp['sm'] = SMdat['SWC1']
    
    data = data.append(tmp)
```
