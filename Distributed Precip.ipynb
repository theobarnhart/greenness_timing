# Distribute Como Creek Precip via Elevation

This notebook looks at how snowmelt is partitioned in Como Creek using the Niwot snotel, AFX ET, and runoff from the como creek gauge.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import pylab
import matplotlib.pyplot as plt
import scipy as sp
import glob
from mpl_toolkits.mplot3d import Axes3D

from scipy import stats

def linpred(slope,intercept,xvals):
    yvals = (xvals*slope)+intercept
    return yvals


pylab.rcParams['figure.figsize'] = (22.0, 11.0)
```

```python
years = range(2005,2013) # generate water years

with np.load('/RHESSys/Comocreek/GRASScomo/barnhatb/cell/como_dem.npz') as data:
    dem = data['arr_0']
```

```python
plt.imshow(dem)
cb = plt.colorbar();
cb.set_label('Elevation [m]', fontsize = 18)
```

DEM loaded sucessfully!

```python
distP = pd.DataFrame(np.zeros([len(years),1]), columns=['wyP'], index=years) # preallocate

eqs = np.zeros([len(years),3]) # preallocate
eqs[:,0] = years
```

```python
eqs[:,1] = [0.537,0.7985,0.6063,0.7032,0.838,0.7861,1.1718,0.7791] # fill in the slopes
```

```python
eqs[:,2] = [-931.22,-1753.2,-1119.9,-1348.5,-1940.7,-1771.9,-2652.8,-1650.1] # fill in the intercepts
```

```python
eqs
```

```python
nans = np.sum(np.isnan(dem))
m,n = np.shape(dem)
area = (m*n) - nans
print (area*(10.*10.)) # print area in m2
```

```python
for i in xrange(0,len(years)):
    prec = linpred(eqs[i,1],eqs[i,2],dem)
    year = eqs[i,0] # get the year
    tot = np.nansum(np.reshape(prec,[1,m*n]))
    distP.loc[years[i],'wyP'] = tot/area # save the arealy averaged precip data
```

This uses the average lapse rate as the lapse rate for 2012, b/c John Knowles' data did not have one.

```python
distP
```

```python
plt.imshow(prec)
cb = plt.colorbar()
cb.set_label('Precipitation [mm]', fontsize = 18)
CS = plt.contour(dem, 70, colors = 'k')
plt.clabel(CS, inline=1, fontsize=10)
plt.title('Como Creek Distributed Precipitation: 2012', fontsize = 24)
plt.ylabel('Relative Northing [x 10 m]', fontsize = 18)
plt.xlabel('Relative Easting [x 10 m]', fontsize = 18);
```
