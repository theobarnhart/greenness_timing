## Calculate Snowpack Metrics

- peak : peakswe [mm]
- date of peak swe : peakswedate
- date of zero snow : snowfree date
- ablation rate : ablrate [mm/day]

```python
years = range(2008,2013) # generate water years
n = len(years)

SWE = pd.DataFrame(np.zeros([n,6]), columns=['peakswe','peakswedate','snowfreedate','ablrate','P','ablseas'], index=years) # preallocate
```

```python
for year in years:
    SWE.loc[year,'peakswe'] = niwot.loc[str(year-1)+'-10-01':str(year)+'-09-30'].swe.max() # extract the peak swe for the water year selected
    SWE.loc[year,'peakswedate'] = niwot.loc[str(year-1)+'-10-01':str(year)+'-09-30'].swe.argmax() # extract the date of peak swe
    SWE.loc[year,'snowfreedate'] = niwot.loc[str(year)+'-01-01':str(year)+'-09-30'].swe.argmin() # extract the first snow free date
    SWE.loc[year, 'P'] = np.sum(wx.loc[str(year-1)+'-10-30':str(year)+'-09-30'].precip)
    
    ## calculate meltout time delta
    timedelta = SWE.loc[year,'snowfreedate'] - SWE.loc[year,'peakswedate']

    SWE.loc[year,'ablrate'] = SWE.loc[year,'peakswe']/timedelta.days # [mm/day]
    SWE.loc[year, 'ablseas'] = int(timedelta.days)
```

## Calculate Total ET and Runoff for water years

```python
SWE['ET'] = np.zeros([n,1]) # preallocate
SWE['Q'] = np.zeros([n,1])
SWE['wy_et'] = np.zeros([n,1])
SWE['wy_q'] = np.zeros([n,1])
```

```python
for year in years:

    # calculate growing season ET from the date of peak swe to the end of the water year.

    strt = SWE.ix[year].peakswedate

    SWE.loc[year,'ET'] = np.sum(ET.loc[str(strt.year)+'-'+str(strt.month)+'-'+str(strt.day):str(year)+'-09-30'].et)
    SWE.loc[year,'wy_et'] = np.sum(ET.loc[str(year-1)+'-10-01':str(year)+'-09-30'].et) # compute water year ET
    
    # calculate the discharge from the date of peak swe to the end of the water year. 

    SWE.loc[year,'Q'] = np.sum(obs.loc[str(strt.year)+'-'+str(strt.month)+'-'+str(strt.day):str(year)+'-09-30'].discharge)
    SWE.loc[year,'wy_q'] = np.sum(obs.loc[str(year-1)+'-10-01':str(year)+'-09-30'].discharge) # compute water year discharge
```

## Calculate traditional snow fraction

```python
n = len(wx)
wx['snow'] = np.zeros([n,1]) # preallocate

for i in xrange(0,n):
    tmpdate = wx.index[i] # get the date of the record for indexing
    if wx.loc[tmpdate,'ATmax'] <= 2:
        wx.loc[tmpdate,'snow'] = wx.loc[tmpdate,'precip']
        
```

```python
for year in years:
    SWE.loc[year,'snow'] = np.sum(wx.loc[str(year-1)+'-10-30':str(year)+'-09-30'].snow)
```

### normalize ET, Q, ans SWE as a fraction of P

```python
SWE['RR'] = SWE.Q/SWE.P #[mm/mm]
SWE['ER'] = SWE.ET/SWE.P #[mm/mm]
SWE['SF'] = SWE.peakswe/SWE.P #[mm/mm]
SWE['TSF'] = SWE.snow/SWE.P #[mm/mm]
```

```python
SWE
```

```python
SWE['DOM'] = np.zeros([len(SWE),1]) # preallocate
SWE['SFD'] = np.zeros([len(SWE),1])
```

```python
# Calculate the day of year of peak SWE and the DOY that the site goes snow free. 
for year in years:
    date = SWE.loc[year,'peakswedate']
    SWE.loc[year,'DOM'] = int(date.strftime('%j'))
    date = SWE.loc[year,'snowfreedate']
    SWE.loc[year,'SFD'] = int(date.strftime('%j'))
```

## Load Distributed P data

## Calculate new metrics

```python
SWE['RR2'] = SWE.Q/distP.wyP
SWE['ER2'] = SWE.ET/distP.wyP
SWE['SF2'] = SWE.peakswe/distP.wyP
SWE['TSF2'] = SWE.snow/distP.wyP
SWE['SF2'] = SWE.peakswe/distP.wyP
```

# Plots and Regression

```python
from scipy import stats

def linpred(slope,intercept,xvals):
    yvals = (xvals*slope)+intercept
    return yvals
```

```python
ttl = 'Distributed Precipiation: peak SWE/P vs Growing Season ET/P'
x = SWE.SF2
xlab = 'Peak SWE/P'
y = SWE.ER2
ylab = 'Evaporative Ratio'

slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
yvals = linpred(slope, intercept, x)


plt.scatter(x, y, 100, c = SWE.index)
plt.colorbar()

plt.title(ttl, fontsize = 22)
plt.ylabel(ylab, fontsize = 16)
plt.xlabel(xlab, fontsize = 16)
plt.plot(x, yvals, 'r-')
print 'pValue: ' + str(p_value)
print 'Slope: ' + str(slope)
print 'Intercept: ' + str(intercept)
```

```python
ttl = 'Distributed Precipiation: Ablation Rate vs Growing Season ET/P'
x = SWE.ablrate
xlab = 'Ablation Rate [mm/day]'
y = SWE.ER2
ylab = 'Evaporative Ratio [1]'

slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
yvals = linpred(slope, intercept, x)


plt.scatter(x, y, 150, c = pd.date_range(start="2008-01-01", end="2012-12-31", freq='A').year)
cb = plt.colorbar()
cb.set_label("Year", fontsize=18)
cb.set_ticks(years)
cb.set_ticklabels(["2008","2009","2010","2011","2012"])

plt.title(ttl, fontsize = 28)
plt.ylabel(ylab, fontsize = 18)
plt.xlabel(xlab, fontsize = 18)
plt.plot(x, yvals, 'r-')
print 'pValue: ' + str(p_value)
print 'Slope: ' + str(slope)
print 'Intercept: ' + str(intercept)

plt.annotate('pValue: ' + str(p_value),xy=[15,0.45], fontsize=18)
plt.annotate('Slope: ' + str(slope),xy=[15,0.44], fontsize=18)
plt.annotate('Intercept: ' + str(intercept),xy=[15,0.43], fontsize=18);
plt.annotate('R$^2$: ' + str(r_value**2),xy=[15,0.42], fontsize=18);

plt.savefig("./ablratevsET_P.png");
```

This is pretty cool. A more rapid ablation rate partitions more water to runoff! Will this hold with distributed ET
