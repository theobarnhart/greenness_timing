```python
%matplotlib inline

import numpy as np
import pandas as pd
import pylab
import matplotlib.pyplot as plt
import scipy as sp
import glob

pylab.rcParams['figure.figsize'] = (22.0, 11.0)
```

```python
niwot = pd.read_pickle('../../data/niwot_snotel.pcl')
```

```python
niwot.swe *= 2.54 # convert from in to cm

niwot.describe()
```

```python
years = np.unique(niwot.index.year)
print 'Calendar Years: '+ str(years)
wyears = years[1:]
print 'Water Years: ' + str(wyears)
```

```python
cols = ['peakswe','peakswedate','snowfreedate','ablationrate','ablationseason', 'growseasonlength']
SWE = pd.DataFrame(np.zeros([len(wyears),len(cols)]), index=wyears, columns=cols)
```

```python
SWE
```

```python
for year in wyears:
    SWE.loc[year,'peakswe'] = niwot.loc[
str(year-1)+'-10-01':str(year)+'-06-20'].swe.max() # extract the peak swe for the water year selected
    
    SWE.loc[year,'peakswedate'] = niwot.loc[
str(year-1)+'-10-01':str(year)+'-06-20'].swe.argmax() # extract the date of peak swe
    
    SWE.loc[year,'snowfreedate'] = niwot.loc[
str(year)+'-01-01':str(year)+'-06-20'].swe.argmin() # extract the first snow free date
    
    timedelta = SWE.loc[year,'snowfreedate'] - SWE.loc[year,'peakswedate']

    SWE.loc[year,'ablationrate'] = SWE.loc[year,'peakswe']/timedelta.days # [mm/day]
    SWE.loc[year, 'ablationseason'] = int(timedelta.days)
    
    timedelta = pd.datetime.strptime(str(year)+'-09-30','%Y-%m-%d') - SWE.loc[year,'peakswedate']
    
    SWE.loc[year,'growseasonlength'] = timedelta.days
```

```python
SWE['snowfreedayofyear'] = np.zeros_like(SWE.ablationseason) # preallocate
SWE['peakswedayofyear'] = np.zeros_like(SWE.ablationseason) # preallocate

for yr in SWE.index:
    SWE.loc[yr,'snowfreedayofyear'] = pd.datetime.strftime(SWE.loc[yr,'snowfreedate'],'%j') # convert date of peak SWE to day of year
    SWE.loc[yr,'peakswedayofyear'] = pd.datetime.strftime(SWE.loc[yr,'peakswedate'],'%j')
```
