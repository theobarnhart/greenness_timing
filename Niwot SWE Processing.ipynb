```python
%matplotlib inline
import numpy as np
import pandas as pd
import pylab
import matplotlib.pyplot as plt
import scipy as sp
import glob
from mpl_toolkits.mplot3d import Axes3D

pylab.rcParams['figure.figsize'] = (22.0, 11.0)
```

## Load snotel data

```python
files = glob.glob('/Users/barnhatb/Dropbox/CUB/ET magnitude and snowmelt/data/snotel/niwot_663/daily/*.csv')
n = len(files)
```

```python
files
```

```python
fl = files[0]
```

```python
niwot = pd.read_csv(fl, skiprows=1, header=True, parse_dates=True, index_col='Date')
#year = int(fl.split('_')[-1].split('=')[-1].split('.')[0])
#niwot.index = niwot.Date
```

```python
for i in xrange(1,n):
    fl = files[i]
    data = pd.read_csv(fl, skiprows=1, header=True, parse_dates=True, index_col='Date')
    niwot = pd.concat([niwot,data])
    
niwot = niwot.sort_index()
```

```python
niwot.head()
```

```python
niwot.tail()
```

```python
cols = ['siteid','time','swe','precip','t','tmax','tmin','tave']

niwot.columns = cols

niwot.swe[niwot.swe < 0] = 0

niwot.swe = (niwot.swe*2.54)*10 # convert in to mm

niwot.precip = (niwot.precip*2.54)*10 # convert in to mm
```

```python
niwot.describe()
```

### SWE

```python
plt.plot(niwot.index, niwot.swe, 'b-', lw =2)
plt.title('Niwot SnoTel Snowpack', fontsize = 28)
plt.xlabel('Time', fontsize = 18)
plt.ylabel('SWE [mm]', fontsize = 18);
#plt.axis('tight');
#plt.savefig('./Niwot_Snopack.png');
```

It looks like low swe years are: 2002, 2004, 2010, 2012

High swe years are: 2003, 2007, 2009, 2011

normal swe years: 2005, 2006, 2008
