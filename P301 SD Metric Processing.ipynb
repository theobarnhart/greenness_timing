```python
niwot = pd.read_pickle('./data/p301/p301_SWE.pcl')
niwot.columns = ['SD_lower','SD_Lflag','SD_upper','SD_Uflag','SWE_lower','swe']
#niwot = pd.DataFrame(niwot, columns=['swe']) # make a data frame with swe even though this is snow depth for compatability with the following script
```

```python
P = pd.read_pickle('./data/p301/p301_P.pcl')
```

```python
years = np.unique(niwot.index.year)
print 'Calendar Years: '+ str(years)
wyears = years[1:]
print 'Water Years: ' + str(wyears)
```

```python
cols = ['peakswe','peakswedate','snowfreedate','ablationrate','ablationseason', 'growseasonlength','ablP']
SWE = pd.DataFrame(np.zeros([len(wyears),len(cols)]), index=wyears, columns=cols)
```

```python
SWE
```

```python
for year in wyears:
    SWE.loc[year,'peakswe'] = niwot.loc[
str(year-1)+'-10-01':str(year)+'-09-30'].swe.max() # extract the peak swe for the water year selected
    
    SWE.loc[year,'peakswedate'] = niwot.loc[
str(year)+'-03-15':str(year)+'-09-15'].swe.argmax() # extract the date of peak swe
    
    SWE.loc[year,'snowfreedate'] = niwot.loc[
str(year)+'-03-15':str(year)+'-09-30'].swe.argmin() # extract the first snow free date
    
    timedelta = SWE.loc[year,'snowfreedate'] - SWE.loc[year,'peakswedate']

    SWE.loc[year,'ablationrate'] = SWE.loc[year,'peakswe']/timedelta.days # [mm/day]
    SWE.loc[year, 'ablationseason'] = int(timedelta.days)
    
    timedelta = pd.datetime.strptime(str(year)+'-09-30','%Y-%m-%d') - SWE.loc[year,'peakswedate']
    
    SWE.loc[year,'growseasonlength'] = timedelta.days
    
    SWE.loc[year,'ablP']= P.loc[SWE.loc[year,'peakswedate']:SWE.loc[year,'snowfreedate']].P_upper.sum()

SWE.ablP /= 10. # convert to cm
SWE['wflux'] = (SWE.peakswe+SWE.ablP)/SWE.ablationseason
```
