```python
%matplotlib inline
import numpy as np
import pandas as pd
import pylab
import matplotlib.pyplot as plt
import scipy as sp
import glob

pylab.rcParams['figure.figsize'] = (22.0, 11.0)
```

```python
cols = ['year','month','day','hour','minute','second','n18sm1','n18sm2','n18sm3','n18sm4','n19sm1','n19sm2','n19sm3','n19sm4']
```

```python
def parser(y,m,d,hh,mm,ss):
    y = int(y)-1
    date = str(m)+"/"+str(d)+"/"+str(y)+" "+str(hh)+":"+str(mm)+":"+str(ss)
    return pd.datetime.strptime(date, '%m/%d/%Y %H:%M:%S')
```

```python
data = pd.read_csv('./data/p301_WB_transect_sm_WY2010.csv',names=cols,
                   parse_dates={'date_time':[0,1,2,3,4,5]},date_parser=parser)
data.index = pd.DatetimeIndex(data.date_time)
del data['date_time']
```
