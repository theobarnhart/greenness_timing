```python
%matplotlib inline
import numpy as np
import pandas as pd
import pylab
import matplotlib.pyplot as plt
import scipy as sp
import glob

pylab.rcParams['figure.figsize'] = (22.0, 11.0)
```

```python
files = glob.glob('../../greenness_timing/snow_depth/P301_WB_transect/wy2011/linear_gap_filled_daily/*.txt')
files
```

files = files[6:8] # grab the two sensors that appear to have a good data record from 2010
files

```python
cols =  ['degC','PAR','mps1','mps2','mps3','mps4','die1','deg1','die2','deg2','die3','deg3','die4','deg4','JtC','Jmm']
```

```python
data = pd.DataFrame() # preallocate

for fl in files:
    
    # grab the sensor name
    tmp = fl.split('/')[-1]
    sens_long = tmp.split('.')[0]
    sens_short = sens_long.split('-')[-1]
    
    # generate a new list of column headers based on the sensor
    cols2 = []
    for name in cols:
        cols2.append(sens_long+'-'+name)
    
    # read in the data
    dat = pd.read_csv(fl)
    dat.index = pd.DatetimeIndex(dat.Date)
    del dat['Date']
    dat.columns = cols2
    
    # append the new data onto the old data frame
    data = pd.concat([data,dat])
    
    del tmp
    del sens_long
    del sens_short
    #del dat
    del cols2
    
```

## Save the Data

```python
data.to_pickle('./data/P301_WB_transect_wy2011.pcl')
```
