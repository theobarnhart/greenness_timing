## Convert LE to ET

Currently this does not support using temperature to change the latent heat of vaporization

```python
# LE in W/m2

def le2et(LE, date):
    
    if all([date.month >= 6, date.month <=8]):  # for summer months
        leofvap = 2500. # latent heat of vaporization J/g
        rho_w = 1000 # kg/m3
    else: # for winter months - all others
        leofvap = 2830. # latent heat of vaporization J/g
        rho_w = 916.7
        
    ET = LE/leofvap # calcualte ET g/m2
    ET /= 1000 # kg/m2
    ET /= rho_w # m/m2
    ET *= 1000 # mm/m2
  
    return ET
```

```python
data['ET'] = le2et(data.LE, data.index)
```

```python
data.ET
```

## Plot Weekly Sums

```python
data = data.interpolate()
data.ET *= (60*30) # expand half hour averages to half hour sums
data2 = data.resample('W', how = 'sum')
```

```python
plot(data2.index, data2.ET, 'b-')
plt.axhline(c='k')
plt.title('T-Van Weekly ET Sums', fontsize = 22)
plt.xlabel('Time', fontsize = 18)
plt.ylabel('ET [mm/m$^2$]', fontsize = 18)
```
