```python
def parser(m,d,y,hhmm):
    if len(hhmm) < 4:
        hhmm = int('0'+str(hhmm))
    hh = hhmm[0:2]
    mm = hhmm[2:4]
    
    date = str(m)+"/"+str(d)+"/"+str(y)+" "+str(hh)+":"+str(mm)
    return pd.datetime.strptime(date, '%m/%d/%Y %H:%M')
```

## Convert to metric

```python
wx.snow_depth *= 2.54
wx.precip *= 2.54

wx.air_Tave -= 32 
wx.air_Tmax -= 32
wx.air_Tmin -= 32

wx.air_Tave /= 1.8 
wx.air_Tmax /= 1.8
wx.air_Tmin /= 1.8
```

```python
wx.describe()
```

```python
plt.plot(wx.index, wx);
```

```python
d = wx[np.isnan(wx.snow_depth)==0].snow_depth.copy()
d[d<0]=0.

swe = wx[np.isnan(wx.SWE)==0].SWE.copy()
swe[swe<0]=0.

swe /= 1000 # conver to m
d /= 1000 # conver to m

d[d<0]=0
rhow = 1000 # density of water [kg/m3]

mw = swe * rhow # compute the mass of the water [kg]

rhos = mw/d # compute the snow density [kg/m3]
rhos[rhos>1000] = 1000.

rhos = rhos.resample('D',how=mean)

wx['snow_density'] = rhos # compute the snow density [kg/m3]
#rhos[rhos>1000] = 1000.
```

```python
for wy in xrange(2009,2013):
    y = wy-1
    
    strt = str(y)+'-12-01'
    nd = str(wy)+'-09-30'
    
    dat=wx[strt:nd].copy()
    
    mxdt = str(dat.snow_density.idxmax())
    
    nd = str(wy)+'-07-28'
    wx.loc[mxdt:nd,'snow_density'] = np.NaN
    dat.loc[mxdt:nd,'snow_density'] = np.NaN
    
    plt.plot(dat.index,dat.snow_density,'ob-');
    plt.ylim([-100,1100]);
    plt.axhline(0);
    plt.title('WY '+str(wy), fontsize=18)
    plt.show()
```

```python
wx.loc[wx.snow_density >=1000,'snow_density'] = np.nan
wx.snow_density = wx.snow_density.interpolate(limit=1)
plt.plot(wx.index,wx.snow_density,'-ob')
```
