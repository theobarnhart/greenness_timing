```python
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
niwot = pd.read_pickle('/RHESSys/Comocreek/barnhatb/clim/niwot_snotel.pcl')
niwot.precip_accum = niwot.precip_accum*25.4
niwot.head()
```

```python
c1 = pd.read_pickle('./C1_chart_prec.pcl')

for i in xrange(0,len(c1)):
    
    # skip entries that are only one day
    if (c1.ix[i].Qual_days == 1) or (np.isnan(c1.ix[i].Qual_days) == True):
        continue
    else:
        days = c1.ix[i].Qual_days
        strt = int(i-(days-1))
        c1.ix[strt:i+1].P = c1.ix[i].P/float(days)
        c1.ix[strt:i+1].Qual_days = 1.

c1.head()
```

```python
afx = pd.read_pickle('./ameriflux/niwot/AFX_niwot.pcl')
afx.head()
```

```python
crn = pd.read_pickle('./NOAA_CRN/boulder_co/boulder_co.pcl')
crn = crn.sort()
crn = crn.interpolate()
crn.tail()
```

## Subset only WY 2008

```python
strt = '2007-10-01'
nd = '2008-09-30'
```

```python
afx = afx[strt:nd]
afx = afx.resample('D', how=np.nansum)
c1 = c1[strt:nd]
niwot = niwot[strt:nd]
crn = crn[strt:nd]
```

```python
afx['accum_prec'] = np.zeros([len(afx),1])
c1['accum_prec'] = np.zeros([len(afx),1])
crn['accum_prec'] = np.zeros([len(afx),1])
for i in xrange(1,len(afx)): 
    afx.accum_prec[i] = afx.accum_prec[i-1] + afx.PREC[i]
    c1.accum_prec[i] = c1.accum_prec[i-1] + c1.P[i]
    crn.accum_prec[i] = crn.accum_prec[i-1] + crn.P[i]
```

```python
crn.P[crn.P == np.NaN]
```

```python
plt.figure(num=None, figsize = (15,8), dpi = 300);
plt.plot(afx.index, afx.accum_prec, lw = 2)
plt.plot(c1.index, c1.accum_prec, lw = 2)
plt.plot(niwot.index, niwot.precip_accum, lw = 2)
plt.plot(crn.index, crn.accum_prec, lw=2);
plt.legend(['AmeriFlux','C1','Niwot SNOTEL', 'NOAA CRN'], loc = 'upper left', fontsize = 16);
plt.title('Como Creek Precipitation Gauges',fontsize = 24)
ax = plt.gca()
plt.setp(ax.get_xticklabels(), fontsize=14);
plt.setp(ax.get_yticklabels(), fontsize=14);
plt.xticks(rotation = -45);
plt.savefig('./figures/WY2008_P.pdf', transparent = True, bbox_inches = 'tight')
```
